```python

a, b, c, d, e = map(int, input("请输入5个数字：").split())
num = [a, b, c, d, e]

num_max = []

def my_max(x,y):
    return x if x>y else y
 
while num:
    m = num[0]
    for i in num[1:]:
        m = my_max(m,i)
    num_max.append(m)
    num.remove(m)
    
a, b, c, d, e = num_max
print(num)
print(num_max)
print(a)
```

```python
digits = [1, 2, 3, 4]
ans = []
for a in digits:
    for b in digits:
        for c in digits:
            if a != b and a != c and b != c:
                ans.append(100*a + 10*b + c)

print(len(ans), sorted(ans))  # 24, [...]
```

```python
import itertools
digits = [1, 2, 3, 4]
ans = [100*a + 10*b + c for a, b, c in itertools.permutations(digits, 3)]
print(len(ans), sorted(ans))  # 24, [...]
```
